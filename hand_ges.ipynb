```python
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.preprocessing.image import ImageDataGenerator

# Define the path to your data folder
data_folder = "D:\\Aptitude Hub\\L\\TK78755\\tk78755-sign-language-recognition-system-using-cnn-and-opencv-V1\\tk78755-sign-language-recognition-system-using-cnn-and-opencv-V1\\Backend\\Gesture Image Data\\"
sign=[]

# Define the gestures you have
gestures = ['A', 'R', '3', 'T', '8', '5', 'W', 'Z', '7', 'P', '1', 'S', '0', 'G', 'E', 'H', 'O', 'U', '9', 'K', 'L', 'N',
            '2', '4', 'C', 'J', '6', 'F', 'Y', 'Q', 'D', 'X', 'M', 'I', 'B', 'V']

# Load and preprocess the images
images = []
labels = []

for gesture in gestures:
    gesture_path = os.path.join(data_folder, gesture)
    for image_name in os.listdir(gesture_path):
        image_path = os.path.join(gesture_path, image_name)
        image = cv2.imread(image_path)
        image = cv2.resize(image, (32, 32))  # Resize the image to a fixed size
        images.append(image)
        labels.append(gestures.index(gesture))

# Convert the lists to numpy arrays
images = np.array(images)
labels = np.array(labels)

# Split the data into training and testing sets
train_images, test_images, train_labels, test_labels = train_test_split(images, labels, test_size=0.2, random_state=42)

# Normalize the image data
train_images = train_images / 255.0
test_images = test_images / 255.0

# Apply data augmentation
datagen = ImageDataGenerator(rotation_range=20, width_shift_range=0.1, height_shift_range=0.1, shear_range=0.1, zoom_range=0.1, horizontal_flip=True)
datagen.fit(train_images)

# Build the CNN model
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(gestures), activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(datagen.flow(train_images, train_labels, batch_size=32), epochs=20, validation_data=(test_images, test_labels))

# Evaluate the model
test_loss, test_accuracy = model.evaluate(test_images, test_labels)
print('Test Loss:', test_loss)
print('Test Accuracy:', test_accuracy)
```

```python
import cv2
import numpy as np
from keras.models import load_model

# Load the trained model
model = load_model('D:\\hand_gesture_cnn_model.h5')

# Define the mapping of class indices to gesture labels
gestures = ['A', 'R', '3', 'T', '8', '5', 'W', 'Z', '7', 'P', '1', 'S', '0', 'G', 'E', 'H', 'O', 'U', '9', 'K', 'L', 'N',
            '2', '4', 'C', 'J', '6', 'F', 'Y', 'Q', 'D', 'X', 'M', 'I', 'B', '_', 'V']

# Initialize the webcam
cap = cv2.VideoCapture(0)

# Check if the webcam is opened successfully
if not cap.isOpened():
    print("Failed to open the webcam")
    exit()

# Set initial predicted gesture as None
predicted_gesture = None

while True:
    # Capture a frame from the webcam
    ret, frame = cap.read()

    # Check if the frame is captured successfully
    if not ret:
        print("Failed to capture the frame")
        break

    # Preprocess the captured frame
    frame_resized = cv2.resize(frame, (32, 32))  # Resize the frame to match the model's input size
    test_image = frame_resized / 255.0  # Normalize the pixel values
    test_image = np.expand_dims(test_image, axis=0)  # Add a batch dimension

    # Make a prediction on the frame
    predictions = model.predict(test_image)
    predicted_class = np.argmax(predictions)

    # Update the predicted gesture only if it has changed
    if predicted_gesture != gestures[predicted_class]:
        predicted_gesture = gestures[predicted_class]
        print("Predicted Gesture:", predicted_gesture)

    # Display the frame with the predicted gesture label
    cv2.putText(frame, "Predicted Gesture: " + predicted_gesture, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    cv2.imshow("Webcam Feed", frame)

    # Exit the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close any open windows
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from keras.models import load_model
import time

# Load the trained model
model = load_model('D:\\hand_gesture_cnn_model.h5')

# Define the lower and upper bounds for the skin color in HSV color space
lower_skin = np.array([0, 20, 70], dtype=np.uint8)
upper_skin = np.array([20, 255, 255], dtype=np.uint8)

# Define the gesture labels
gestures = ['A', 'R', '3', 'T', '8', '5', 'W', 'Z', '7', 'P', '1', 'S', '0', 'G', 'E', 'H', 'O', 'U', '9', 'K', 'L', 'N',
            '2', '4', 'C', 'J', '6', 'F', 'Y', 'Q', 'D', 'X', 'M', 'I', 'B', '_', 'V']

# Initialize the webcam
cap = cv2.VideoCapture(0)

# Initialize the delay timer
delay_timer = 0

while True:
    # Read the frame from the webcam
    ret, frame = cap.read()

    # Check if delay timer has expired
    if time.time() > delay_timer:
        # Convert the frame to the HSV color space
        hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

        # Create a binary mask of the pixels that fall within the skin color range
        skin_mask = cv2.inRange(hsv_frame, lower_skin, upper_skin)

        # Perform morphological operations to remove noise and enhance the mask
        kernel = np.ones((5, 5), np.uint8)
        skin_mask = cv2.morphologyEx(skin_mask, cv2.MORPH_OPEN, kernel)
        skin_mask = cv2.dilate(skin_mask, kernel, iterations=4)

        # Find contours of the hand region
        contours, hierarchy = cv2.findContours(skin_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        # Find the largest contour (assumed to be the hand)
        if len(contours) > 0:
            hand_contour = max(contours, key=cv2.contourArea)

            # Draw a bounding rectangle around the hand contour
            x, y, w, h = cv2.boundingRect(hand_contour)
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

            # Extract the hand region from the frame
            hand_region = frame[y:y + h, x:x + w]

            # Preprocess the hand image for prediction
            hand_image = cv2.resize(hand_region, (32, 32))
            hand_image = hand_image / 255.0
            hand_image = np.expand_dims(hand_image, axis=0)

            # Perform the prediction
            predictions = model.predict(hand_image)
            predicted_class = np.argmax(predictions)
            predicted_gesture = gestures[predicted_class]

            # Display the predicted gesture on the frame
            cv2.putText(frame, 'Gesture: ' + predicted_gesture, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

            # Set the delay timer for the next prediction
            delay_timer = time.time() + 1  # 
```
